```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns;sns.set()
```

```python
from sklearn.datasets.samples_generator import make_blobs
X,y = make_blobs(n_samples=50,centers=2,random_state=0,cluster_std=0.6)
plt.scatter(X[:,0],X[:,1],c=y,s=50,cmap='autumn')
```

```python
xfit = np.linspace(-1,3.5)
plt.scatter(X[:,0],X[:,1],c=y,s=50,cmap='autumn')
plt.plot([0.6],[2.1],'X',color='red',markeredgewidth=2,markersize=0.2)

for m,b in [(1,0.65),(0.5,1.6),(-0.2,2.9)]:
    plt.plot(xfit,m * xfit + b,'-k')
plt.xlim(-1,3.5)
```

```python
xfit = np.linspace(-1,3.5)
plt.scatter(X[:,0],X[:,1],c=y,s=50,cmap='autumn')

for m,b,d in [(1,0.65,0.33),(0.5,1.6,0.55),(-0.2,2.9,0.2)]:
    yfit = m * xfit + b
    plt.plot(xfit,yfit,'-k')
    plt.fill_between(xfit,yfit - d,yfit + d,edgecolor='none',color='#AAAAAA',alpha=0.4)
    plt.xlim(-1,3.5)
```

```python
from sklearn.svm import SVC #"Support vector classifier"支持向量分类器
model = SVC(kernel='linear',C=1E10)
model.fit(X,y)
```

```python
def plot_svc_decision(model,ax=None,plot_support=True):
    """画二维SVC的决策函数"""
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()

    #创建评估模型的网络
    x = np.linspace(xlim[0],xlim[1],30)
    print("x的值是：",x.shape)
    y = np.linspace(xlim[0],ylim[1],30)
    print("y的值是：",y.shape)
    Y,X = np.meshgrid(y,x)
    print("Y的值是：",Y.shape)
    print("X的值是：",X.shape)
    xy = np.vstack([X.ravel(),Y.ravel()]).T
    print("xy的值是：",xy.shape)
    P = model.decision_function(xy).reshape(X.shape)  #计算样本点到分割超平面的距离
    print("P的值是：",P.shape)
    #画决策边界和边界
    ax.contour(X,Y,P,colors='k',levels=[-1,0,1],alpha=0.5,linestyles=['--','-','--'])

    #画支持向量
    if plot_support:
        print(model.support_vectors_)
        ax.scatter(model.support_vectors_[:,0],model.support_vectors_[:,1],
                  s=300,linewidth=1,facecolors='none');
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
```

```python
plt.scatter(X[:,0],X[:,1],c=y,s=50,cmap='autumn')
plot_svc_decision(model);
```
